# Imports

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

# Setup imaging system

```python
dim = 500
x_ax, y_ax = np.linspace(-1, 1, dim+1), np.linspace(-1, 1, dim+1)
x, y = np.meshgrid(x_ax, y_ax)
A = np.zeros((dim+1, dim+1))
A[(x**2 + y**2) < 1] = 1
alpha, beta, phi = np.zeros_like(A), np.zeros_like(A), np.zeros_like(A)
E_cor = A * np.exp(alpha + 1j*beta) * np.exp(1j * phi)
# E_cor = np.fft.fft2(A_cor)
I_cor = np.absolute(E_cor)**2.
I_im = np.absolute(np.fft.fft2(I_cor))

plt.figure(figsize=(8,8))
plt.pcolormesh(x_ax, y_ax, I_cor)
plt.show()

plt.figure(figsize=(8,8))
plt.pcolormesh(x_ax, y_ax, I_im, norm=matplotlib.colors.LogNorm())
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.imshow(I_im, norm=matplotlib.colors.LogNorm())
plt.show()
```
