# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy.signal import argrelextrema
```

# X direction

### Throughput vs. angular separation

1) Read in data.
2) Section data into trials.
3) Fit (quadratic) for center of peaks.
4) Plot profiles.

```python
# Read in data
data_x = pd.read_csv('VFN_demo_csv_x_220428.csv')

# Section data into trials
lower, = np.where(np.array(data_x['x_ticks']) == np.array(data_x['x_ticks']).min())
upper, = np.where(np.array(data_x['x_ticks']) == np.array(data_x['x_ticks']).max())
trials = []
for i in range(len(upper)):
    trials.append(np.array(data_x)[upper[i]:lower[i]])
    if i+1 < len(upper):
        trials.append(np.array(data_x)[lower[i]:upper[i+1]])    
        
trials = np.array(trials)
peaks = np.zeros((len(trials),2))

for t in range(len(trials)):
    trial = trials[t]
    indices_max, = argrelextrema(trial[:,1], np.greater)
    for i, index_max in enumerate(indices_max):
        x_fit, power_fit = trial[index_max-4:index_max+5, 0], trial[index_max-4:index_max+5, 1]
        # Quadratic fit to peaks
        coefs = np.polyfit(x_fit, power_fit, 2)
        b = np.poly1d(coefs)
        peaks[t, i] = scipy.optimize.fmin(-b, x_fit[np.argmin(-power_fit)])[0]
        
# Plot profiles
plt.figure(figsize=(10,6))
colors = ['blue', 'red', 'green']
for i, peak in enumerate(peaks):
    plt.plot(trials[i][:,0], trials[i][:,1], 'o-', color=colors[i])
    plt.axvline(peak[0], linestyle='dashed', color=colors[i])
    plt.axvline(peak[1], linestyle='dashed', color=colors[i])
plt.xlabel(r'x (mm)')
plt.ylabel('Power ($\mu$W)')
plt.show()
```

### Compute separation of peaks.

```python
# Convert to angular separation
f = 20 # mm
D = 4.5 # mm
F_num = f/D # F-number
lambda_source = 635 * 10**6/(10**9) # laser wavelength in mm
x_sep = abs(np.diff(peaks, axis=1).flatten())/2.
x_0 = x_sep + np.min(peaks, axis=1)
theta_sep = x_sep/(lambda_source * F_num)
theta_peaks = []
thetas = []
for t, trial in enumerate(trials):
    thetas.append((trial[:, 0]-x_0[t])/(lambda_source * F_num))
    theta_peaks.append((peaks[t]-x_0[t])/(lambda_source * F_num))

# Compute separation of peaks for each trial
print('Peak separations for each trial (in lambda/D): \n', theta_sep)

# Plot profiles
plt.figure(figsize=(10,6))
colors = ['blue', 'red', 'green']
for i, theta_peak in enumerate(theta_peaks):
    plt.plot(thetas[i], trials[i][:,1], 'o-', color=colors[i], label = 'Trial '+str(i))
    plt.axvline(theta_peak[0], linestyle='dashed', color=colors[i])
    plt.axvline(theta_peak[1], linestyle='dashed', color=colors[i])
plt.xlabel(r'$\theta$ ($\lambda$/D)')
plt.ylabel('Power ($\mu$W)')
plt.legend()
plt.show()
```

# Y direction

### Throughput vs. angular separation

```python
# Read in data
data_x = pd.read_csv('VFN_demo_csv_y_220428.csv')

# Section data into trials
lower, = np.where(np.array(data_x['y_ticks']) == np.array(data_x['y_ticks']).min())
upper, = np.where(np.array(data_x['y_ticks']) == np.array(data_x['y_ticks']).max())
trials = []
for i in range(len(upper)):
    trials.append(np.array(data_x)[upper[i]:lower[i]])
    if i+1 < len(upper):
        trials.append(np.array(data_x)[lower[i]:upper[i+1]])    
        
trials = np.array(trials)
peaks = np.zeros((len(trials),2))

for t in range(len(trials)):
    trial = trials[t]
    indices_max, = argrelextrema(trial[:,1], np.greater)
    is_max, = np.where(trial[:,1][indices_max] > 0.1)
    indices_max = indices_max[is_max]
    for i, index_max in enumerate(indices_max):
        x_fit, power_fit = trial[index_max-4:index_max+5, 0], trial[index_max-4:index_max+5, 1]
        # Quadratic fit to peaks
        coefs = np.polyfit(x_fit, power_fit, 2)
        b = np.poly1d(coefs)
        peaks[t, i] = scipy.optimize.fmin(-b, x_fit[np.argmin(-power_fit)])[0]
        
# Plot profiles
plt.figure(figsize=(10,6))
colors = ['blue', 'red', 'green']
for i, peak in enumerate(peaks):
    plt.plot(trials[i][:,0], trials[i][:,1], 'o-', color=colors[i])
    plt.axvline(peak[0], linestyle='dashed', color=colors[i])
    plt.axvline(peak[1], linestyle='dashed', color=colors[i])
plt.xlabel(r'y (mm)')
plt.ylabel('Power ($\mu$W)')
plt.show()
```

### Compute separation of peaks.

```python
# Convert to angular separation
f = 20 # mm
D = 4.5 # mm
F_num = f/D # F-number
lambda_source = 635 * 10**6/(10**9) # laser wavelength in mm
x_sep = abs(np.diff(peaks, axis=1).flatten())/2.
x_0 = x_sep + np.min(peaks, axis=1)
theta_sep = x_sep/(lambda_source * F_num)
theta_peaks = []
thetas = []
for t, trial in enumerate(trials):
    thetas.append((trial[:, 0]-x_0[t])/(lambda_source * F_num))
    theta_peaks.append((peaks[t]-x_0[t])/(lambda_source * F_num))

# Compute separation of peaks for each trial
print('Peak separations for each trial (in lambda/D): \n', theta_sep)

# Plot profiles
plt.figure(figsize=(10,6))
colors = ['blue', 'red', 'green']
for i, theta_peak in enumerate(theta_peaks):
    plt.plot(thetas[i], trials[i][:,1], 'o-', color=colors[i], label = 'Trial '+str(i))
    plt.axvline(theta_peak[0], linestyle='dashed', color=colors[i])
    plt.axvline(theta_peak[1], linestyle='dashed', color=colors[i])
plt.xlabel(r'$\theta$ ($\lambda$/D)')
plt.ylabel('Power ($\mu$W)')
plt.legend()
plt.show()
```
